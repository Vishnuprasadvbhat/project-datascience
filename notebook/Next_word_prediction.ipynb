```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([data])
```

```python
tokenizer.word_counts
```

```python
tokenizer.word_index
```

```python
input_seq = []

for sentences in data.split('\n'):
  # print(sentences)
  # print(tokenizer.texts_to_sequences([sentences])[0])
  tokenized_sent= tokenizer.texts_to_sequences([sentences])[0]

  for i in range(1,len(tokenized_sent)):
    input_seq.append(tokenized_sent[:i+1])
```

```python
input_seq
```

```python
# the size of the rows are not same, when we perform supevised leanring we need same number of rows

# we perform zero padding

# Before that we need to know max length of rows

max_len = max([len(x) for x in input_seq])
print(max_len)
```

```python
from tensorflow.keras.preprocessing.sequence  import pad_sequences
```

```python
padded_input_sequences = pad_sequences(input_seq,maxlen=max_len,padding='pre')
```

```python
padded_input_sequences
```

```python
x = padded_input_sequences[:,:-1]
x
```

```python
y = padded_input_sequences[:,-1]
y
```

```python
x.shape
```

```python
y.shape
```

```python
len(tokenizer.word_index)
tokenizer.word_index
```

```python
from tensorflow.keras.utils import to_categorical

y_cat= to_categorical(y,num_classes =389)
```

```python
y
```

```python
y_cat.shape
```

# Now Start with building the model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense
```

```python
model = Sequential()
model.add(Embedding(389,50, input_length=36))
model.add(LSTM(150))
model.add(Dense(389, activation = 'softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```
